Importar librerias

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
import cv2 as cv
from matplotlib.colors import hsv_to_rgb
```

Imagen

```python
img = cv.imread("bags.png") 
plt.imshow(img)
plt.show()
```

Imagen de BGR a RGB

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## Método gráfica scatter

```python
r, g, b = cv.split(img)
fig = plt.figure()
axis = fig.add_subplot(1,1,1,projection = "3d")
pixel_colors = img.reshape((np.shape(img)[0]*np.shape(img)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
axis.scatter(r.flatten(),g.flatten(),b.flatten(),facecolors = pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

```python
hsv_img = cv.cvtColor(img, cv.COLOR_RGB2HSV)
h,s,v = cv.split(hsv_img)
fig = plt.figure()
axis = fig.add_subplot(1,1,1,projection="3d")
axis.scatter(h.flatten(),s.flatten(),v.flatten(), facecolors = pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Saturation")
axis.set_zlabel("Value")
plt.show()
```

#### Color Azul

```python
light_blue = (300,255,255)
dark_blue = (100,127,50)

lb_square = np.full((10,10,3), light_blue,dtype=np.uint8)/255.0
db_square = np.full((10,10,3), dark_blue, dtype=np.uint8)/255.0
plt.subplot(2,2,1)
plt.imshow(hsv_to_rgb(lb_square))
plt.subplot(2,2,2)
plt.imshow(hsv_to_rgb(db_square))
plt.show()
```

##### Máscara azul

```python
mask = cv.inRange(hsv_img,dark_blue,light_blue)
result_blue = cv.bitwise_and(img,img,mask=mask)
plt.subplot(1,2,1)
plt.imshow(mask,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(result_blue)
plt.show()
```

Importar librerias faltantes

```python
import skimage
from skimage.io import imread, imshow
from skimage.color import rgb2hsv
```

Análisis de *hue*, *transparency* y *value*

```python
hsv_img = rgb2hsv(img)

fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].imshow(hsv_img[:,:,0], cmap='hsv')
ax[0].set_title('hue')

ax[1].imshow(hsv_img[:,:,1], cmap='hsv')
ax[1].set_title('transparency')

ax[2].imshow(hsv_img[:,:,2], cmap='hsv')
ax[2].set_title('value')

fig.colorbar(imshow(hsv_img[:,:,2],cmap='hsv'))
fig.tight_layout()

plt.show()
```

Color azul

```python
#refer to hue channel (in the colorbar)
lower_mask = hsv_img[:,:,0] > 0.5

#refer to hue channel (in the colorbar)
upper_mask = hsv_img[:,:,0] < 0.9

# refer to transparency channel (in the colorbar)
saturation_mask =  hsv_img[:,:,1] > 0.5

mask2 = upper_mask*lower_mask*saturation_mask
red = img[:,:,0]*mask2
green = img[:,:,1]*mask2
blue = img[:,:,2]*mask2
img_masked_blue = np.dstack((red,green,blue))
imshow(img_masked_blue)

plt.show()
```

Color naranja

```python
#refer to hue channel (in the colorbar)
upper_mask = hsv_img[:,:,0] < 0.07

#refer to tranparency channel (in the colorbar)
saturation_mask = hsv_img[:,:,1] > 0.55

#refer to value channel
value_mask = hsv_img[:,:,2] > 0.5

mask2 = upper_mask*saturation_mask*value_mask
red = img[:,:,0]*mask2
green = img[:,:,1]*mask2
blue = img[:,:,2]*mask2
img_masked_orange = np.dstack((red,green,blue))
imshow(img_masked_orange)

plt.show()
```

Color verde

```python
#refer to hue channel (in the colorbar)
lower_mask = hsv_img[:,:,0] > 0.2

#refer to hue channel (in the colorbar)
upper_mask = hsv_img[:,:,0] < 0.6

#refer to tranparency channel (in the colorbar)
saturation_mask = hsv_img[:,:,1] > 0.23

mask2 = upper_mask*lower_mask*saturation_mask
red = img[:,:,0]*mask2
green = img[:,:,1]*mask2
blue = img[:,:,2]*mask2
img_masked_green = np.dstack((red,green,blue))
imshow(img_masked_green)

plt.show()
```

Color amarillo

```python
#refer to hue channel (in the colorbar)
lower_mask = hsv_img[:,:,0] > 0.08

#refer to hue channel (in the colorbar)
upper_mask = hsv_img[:,:,0] < 0.2

#refer to tranparency channel (in the colorbar)
saturation_mask = hsv_img[:,:,1] > 0.57

#refer to value channel
value_mask = hsv_img[:,:,2] > 0.19

mask2 = upper_mask*lower_mask*saturation_mask*value_mask
red = img[:,:,0]*mask2
green = img[:,:,1]*mask2
blue = img[:,:,2]*mask2
img_masked_yellow = np.dstack((red,green,blue))
imshow(img_masked_yellow)

plt.show()
```

Union de *bags*

```python
bags_union = img_masked_blue+img_masked_orange+img_masked_green+img_masked_yellow
imshow(bags_union)
plt.show()
```
